```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import sympy as sp
```

## Problem 0: some fake data to mess with

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

## Problem 1: subsets of data

```python
fig, ax = plt.subplots(1,1,figsize=(11.5,8))
x1=np.arange(1000)
ax.scatter(x1,bkgd[:1000]);
```

```python
#1c
i=100
hunnid=[]
while i<int(1e5):
    hunnid.append(bkgd[i])
    i=i+100
```

```python
fig, ax = plt.subplots(1,1,figsize=(11.5,8))
x3=np.arange(len(hunnid))
ax.scatter(x3,hunnid);
```
