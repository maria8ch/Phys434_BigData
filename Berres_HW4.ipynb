```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import sympy as sp
```

## Problem 0: some fake data to mess with

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

## Problem 1: subsets of data
## 1. a)

```python
fig, ax = plt.subplots(1,1,figsize=(11.5,8))
ax.set_title('First 1000 points')
ax.set_xlabel('time')
ax.set_ylabel('Distribution Points')
ax.scatter(t[:1000],bkgd[:1000]);
```

## 1. b)

```python
def xandy(x,y):
    if x+y>len(bkgd):
        print('WRONG!')
        return
    else:
        bs=bkgd[y:x+y]
        ts=t[y:x+y]
        return bs,ts
```

```python
fig, ax = plt.subplots(1,1,figsize=(11.5,8))
bs,ts=xandy(2000,int(1e5/2))
ax.set_title(f'2000 Points Starting at point {int(1e5/2):0.0f}')
ax.set_xlabel('time')
ax.set_ylabel('Distribution Points')
ax.scatter(ts,bs);
```

## 1. c)

```python
i=100
hunnid=[]
ht=[]
while i<int(1e5):
    hunnid.append(bkgd[i])
    ht.append(t[i])
    i=i+100
```

```python
fig, ax = plt.subplots(1,1,figsize=(11.5,8))
ax.set_title('Every 100th Point')
ax.set_xlabel("time in 100's")
ax.set_ylabel('Distribution Points')
ax.scatter(ht,hunnid);
```

## Problem 2: density
## 2. a)

```python
fig, ax = plt.subplots(1,1,figsize=(11,10))
x4=np.arange(len(bkgd))
ax.set_title('2D Histogram Density Plot')
ax.set_xlabel('time')
ax.set_ylabel('Distribution Points Density')
d=ax.hist2d(t,bkgd,bins=100);
fig.colorbar(d[3]);
```

## 2. b)

We are plotting the amount of counts found in each bin. Since the counts can't be represented as a y-value height like in a 1-Dimensional histogram, we have to use another method to show the value of each bin. This plot represents them by corresponding a color to a relative bin count, as the number of counts in a bin increases the more yellow from blue the color becomes, as the number of counts goes to zero, the more blue from yellow the bins look.

## Problem 3: folding

```python
tf = np.linspace(0,np.pi*100,np.int(1e5)) % (2*np.pi)
Nf = 10+np.sin(tf)
bkgdm = stats.norm.rvs(size=np.int(1e5))*np.sqrt(Nf)+Nf
```

## 3. a)

```python
fig, ax = plt.subplots(1,1,figsize=(11.5,8))
ax.set_title('Folded distribution')
ax.set_xlabel('Folded time')
ax.set_ylabel('Distribution Points')
ax.scatter(tf,bkgdm);
```

## 3. b)

```python
fig, ax = plt.subplots(1,1,figsize=(11,10))
x4=np.arange(len(bkgdm))
ax.set_title('2D Histogram Density Plot Folded')
ax.set_xlabel('Folded time')
ax.set_ylabel('Distribution Points Density')
d=ax.hist2d(tf,bkgdm,bins=100);
fig.colorbar(d[3]);
```
